```python
#importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
%matplotlib inline
sns.set(color_codes=True)
from sklearn.linear_model import LinearRegression
from sklearn import linear_model
```

```python
df = pd.read_excel('/content/India_Data.xlsx')
```

```python
df
```

```python
df.info
```

```python
x = df.iloc[:, :-18]
```

```python
x
```

```python
x.info
```

```python
sheet_name = 1
Y = pd.read_excel('/content/India_Data.xlsx', sheet_name=sheet_name)
```

```python
Y
```

```python
#splitting data into train and test
X_train, X_test, y_train, y_test = train_test_split(x, Y, test_size = 0.2, random_state=99)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
```

```python
rf_model.fit(X_train, y_train)
```

```python
# Make predictions on the test set
y_pred = rf_model.predict(X_test)

# Calculate Mean Squared Error
mse = mean_squared_error(y_test, y_pred)

# Calculate R-squared (R2) score
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"R-squared (R2) Score: {r2}")
```

```python
y_pred = rf_model.predict(X_test)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, alpha=0.5)
plt.xlabel('Actual Adoption cost')
plt.ylabel('Predicted Adoption cost')
plt.title('Actual vs. Adoption cost')
plt.grid(True)
plt.show()
```

```python
# Calculate residuals
residuals = np.array(y_test) - np.array(y_pred)
```

```python
# Box plot of residuals
plt.subplot(2, 2, 3)
sns.boxplot(residuals)
plt.title('Residuals - Box Plot')
```

```python
# Density plot of residuals

plt.figure(figsize=(16, 6))
sns.kdeplot(residuals, shade=True)
plt.xlabel('Residuals')
plt.title('Residuals - Density Plot')

plt.tight_layout()
plt.show()
```

```python
# Regression Line Plot
plt.figure(figsize=(12, 4))

# Scatter plot with a regression line
plt.subplot(1, 3, 1)
sns.regplot(x=y_test, y=y_pred, line_kws={"color": "red"}, scatter_kws={"alpha": 0.5})
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Regression Line Plot')
```

```python
# Lorenz Curve
plt.subplot(1, 3, 2)
y_test_sorted = np.sort(y_test)
y_pred_sorted = np.sort(y_pred)
plt.plot(np.cumsum(y_test_sorted) / np.sum(y_test_sorted), label='Actual')
plt.plot(np.cumsum(y_pred_sorted) / np.sum(y_pred_sorted), label='Predicted', linestyle='--')
plt.xlabel('Percentile')
plt.ylabel('Cumulative Sum')
plt.title('Lorenz Curve')
plt.legend()
```

```python
# QQ Plot
import scipy.stats as stats
import statsmodels.api as sm

sm.qqplot(residuals, line='s')
plt.title('QQ Plot')

plt.tight_layout()
plt.show()
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
y_pred = model.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print(f"MAE: {mae}")
print(f"MSE: {mse}")
print(f"RMSE: {rmse}")
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, alpha=0.5)
plt.xlabel('Actual Adoption cost')
plt.ylabel('Predicted Adoption cost')
plt.title('Actual vs. Adoption cost')
plt.grid(True)
plt.show()
```

```python
y_pred
```

# LSTM

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
train_scaled = scaler.fit_transform(X_train)
test_scaled = scaler.fit_transform(X_test)
```

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM, Dropout

model = Sequential([
    LSTM(64, return_sequences=True, input_shape=(X_train.shape[1], 1)),
    Dropout(0.2),
    LSTM(64),
    Dropout(0.2),
    Dense(32, activation='relu'),
    Dropout(0.2),
    Dense(1)
])
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
# Train the model
history = model.fit(train_scaled, y_train, validation_data=(test_scaled, y_test), epochs=10, batch_size=64)
```

```python
test_loss = model.evaluate(test_scaled, y_test)
print(f"Test Loss: {test_loss}")
```

```python
predictions = model.predict(test_scaled)
```

```python
mae = mean_absolute_error(y_test, predictions)
print(f"Mean Absolute Error (MAE): {mae}")
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(y_test, predictions)
plt.xlabel("Actual Cost")
plt.ylabel("Predicted Cost")
plt.title("Actual vs. Predicted Cost")
plt.show()
```

```python
predictions
```
